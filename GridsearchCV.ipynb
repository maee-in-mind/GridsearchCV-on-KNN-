```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset=pd.read_csv("diabetes.csv")
dataset.head()
```

```python
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
x.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20)
```

```python
model11=knn(n_neighbors=2,weights='distance', algorithm='auto',
         leaf_size=10,p=1)
model11.fit(x_train,y_train)
y_predict11=model11.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_predict11,y_test)*100
```

```python
#now use GridSearchCV for hyperparamerter tuning
from sklearn.model_selection import GridSearchCV
parameters={'n_neighbors':[2,3,4,5,6,7],'weights':['uniform','distance'],
            'algorithm':['auto','ball_tree','kd_tree','brute'],'p':[1,2],
           'leaf_size':[10,20,30,40,50],'metric':['minkowski']
            
            }



grid = GridSearchCV(model11, parameters, refit = True, verbose = 3) 

  
# fitting the model for grid search 
grid.fit(x_train, y_train) 
```

```python
grid_predictions = grid.predict(x_test)
```

```python
accuracy_score(grid_predictions,y_test)*100
```

```python
print(grid.best_params_)
```
